```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import math
from Unsupervised.MBSAS import MBSAS
```

```python
iris = datasets.load_iris()

s = plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target, cmap='Set2')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend(s.legend_elements()[0], iris.target_names)
plt.title('Sepal Graph')
plt.show()
```

```python
s = plt.scatter(iris.data[:, 2], iris.data[:, 3], c=iris.target, cmap='Set2')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.legend(s.legend_elements()[0], iris.target_names)
plt.title('Petal Graph')
plt.show()
```

```python
train = iris.data[:140, 2:4]
test = iris.data[140:151, 2:4]
```

```python
# Basic Sequential Algorithmic Scheme (BSAS) Implementation
mbsas = MBSAS()
thetas, ks = mbsas.optimize(train)

print(thetas[-1])
print(ks[-1])

plt.plot(thetas, ks)
plt.xlabel('thetas')
plt.ylabel('number of clusters')
plt.show()
```

```python
avg, k = mbsas.pick_hyperparams(train)
c = mbsas.fit(train, 3, 2)
s = plt.scatter(train[:, 0], train[:, 1], c=c, cmap='Set2')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.legend(s.legend_elements()[0], iris.target_names)
plt.title('Petal Graph')
plt.show()
```

```python
c = mbsas.predict(test)
s = plt.scatter(test[:, 0], test[:, 1], c=c, cmap='Set2')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.legend(s.legend_elements()[0], iris.target_names)
plt.title('Petal Graph')
plt.show()
```
